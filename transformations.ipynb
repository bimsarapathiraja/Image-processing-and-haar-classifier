# Scaling

```python
import cv2 as cv
import numpy as np
```

```python
img = cv.imread('avengers.jpg', 1)
```

```python
height, width = img.shape[:2]
print(height, width)
```

```python
res = cv.resize(img, None, fx = 2, fy = 2, interpolation = cv.INTER_CUBIC)

res2 = cv.resize(img, (2*width, 2*height), interpolation = cv.INTER_CUBIC)
```

```python
cv.imshow('image', img)
cv.imshow('res', res)
cv.imshow('res1', res2)

cv.waitKey(0)

cv.destroyAllWindows()
```

# Translation

```python
img1 = cv.imread('avengers.jpg', 0)
rows,cols = img1.shape
```

```python
M = np.float32([[1,0,100], [0, 1, 50]])
```

```python
dst = cv.warpAffine(img1, M, (cols, rows))
```

```python
cv.imshow('img', dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Rotation

```python
MM = cv.getRotationMatrix2D((cols/2, rows/2), 90, 1)
```

```python
dst = cv.warpAffine(img, MM, (cols, rows))
```

```python
cv.imshow('img', dst)
cv.imshow('img2', img)
cv.waitKey(3000)
cv.destroyAllWindows()
```

# Affine Transformation

```python
import matplotlib.pyplot as plt
```

```python
rows, cols, ch= img.shape
```

```python
pts1 = np.float32([[50,50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])
```

```python
M3 = cv.getAffineTransform(pts1, pts2)
```

```python
dst = cv.warpAffine(img, M3, (cols, rows))
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Input')
plt.subplot(122), plt.imshow(dst), plt.title('Output')
plt.show()
```

# Perspective transformation

```python
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
```

```python
M = cv.getPerspectiveTransform(pts1, pts2)
```

```python
dst = cv.warpPerspective(img, M, (300, 300))
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Input')
plt.subplot(122), plt.imshow(dst), plt.title('Output')
```
