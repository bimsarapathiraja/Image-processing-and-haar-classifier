```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('avengers.jpg')
```

```python
kernel = np.ones((5,5), np.float32)/25
```

```python
dst = cv2.filter2D(img, -1, kernel)
```

```python
plt.subplot(121), plt.imshow(img), plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dst), plt.title('filtred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cv2.imshow('dst', dst)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Blur

```python
img = cv2.imread('noise.png')
blurAvg = cv2.blur(img, (5,5))
blurGau = cv2.GaussianBlur(img, (5,5), 0)
blurMed = cv2.medianBlur(img, 5)
blurBil = cv2.bilateralFilter(img, 9, 75, 75)
```

```python
cv2.imshow('dst', blurAvg)
cv2.imshow('dst1', blurGau)
cv2.imshow('dst2', blurMed)
cv2.imshow('dst3', blurBil)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
